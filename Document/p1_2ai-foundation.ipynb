## Python基本语法

```python
# 变量定义（赋值语句：变量名 = 值）
a = 1
b = 2
print(a,b)   # 内置函数print(),输出内容到终端
```

```python
# 基本运算
c = a + b
print(c)
```

```python
# 计算BMI
weight = 70
height = 1.75
bmi = weight/height**2
print(f'BMI={bmi}')
```

```python
# 数据类型
# 数值型：整形-int，浮点型-float，复数-complex
n = 3-5j
print(type(weight),type(bmi),type(n))
```

```python
# 字符串-str，布尔型-bool（True/False）
name = '张三'
flag = False
print(type(name),type(flag))
```

## 列表-list
Python列表使用方括号定义，并通过逗号分隔元素。  
列表中包含的元素可以是各种类型的数据，例如数字、字符串、甚至其他列表。

```python
# 列表定义-[]
l1 = [1,2,3,3.4,4]
print(l1, type(l1))
```

```python
# 列表创建-list(string)
l2 = list('杭州师范大学')
print(l2)
```

```python
# 列表生成-列表推导式
l3 = [item+10 for item in l1]
print(l3, type(l3))
```

```python
# 列表操作
scores = [85,92,78,90,88]
scores.append(95)
scores.sort()
print(type(scores), scores)
```

## 字典-dict
Python字典使用花括号定义，是一组键-值对集合，可通过"键"快速访问"值"。

```python
student = {'name':'李明','age':18,'major':'国际商务'}
print(student['name'])
```

```python
student['age'] = 20
student['grade'] = 'A'  
print(type(student), student)
```

## 元组-tuple
Python元组使用圆括号定义，是不可变序列，一旦创建不可修改。

```python
dimensions = (1920,1080)
print(dimensions[0])
# dimensions[0] = 1800
```

## 索引和切片：字符串、列表、元组序列类型都支持索引和切片
索引从0开始，支持负索引（末尾向前数）  
切片允许提取序列的一部分，[start:end]含start，不含end

```python
numbers = [10,20,30,40,50]
first = numbers[0]
last = numbers[-1]
sub_numbers = numbers[1:4]
print(first, last, sub_numbers)
```

```python
reverse_numbers = numbers[::-1]
print(reverse_numbers)
```

```python
n1 = numbers[1::2]
print(n1)
n2 = numbers[-2:0:-1]
print(n2)
```

## 控制流-条件语句：if-elif-else

```python
## 情感分类
def classify_sentiment(score):
    if score > 0.7:
        return 'positive'
    elif score < 0.3:
        return 'negative'
    else:
        return 'neutral'
```

```python
print(classify_sentiment(0.8))
```

## 控制流-条件语句：for/while

```python
# 求1-100的和
sum = 0
for num in range(1,101):
    sum += num
print(f'1-100的和为{sum}')    
```

```python
sum = 0
num = 1
while num<=100:
    sum += num
    num = num + 1
print(f'1-100的和为{sum}')    
```

## 模块引入-import
import 包[.子包]  
import 包[.子包].模块  
from 包[.子包] import 模块  
from 包[.子包].模块 import 函数/类/变量  
**注意：**  
1.import 的最后一段必须是模块或包，不能是函数/类/变量  
2.想直接拿到函数/类/变量，就用 from ... import ...

```python
# 引入Python标准库中的random模块
import random
m = random.random()
print(m)
```

```python
# 引入random模块的random()函数
from random import random
n = random()
print(n)
```

```python
for i in range(6):
    m_i = random()
    print(f'm_{i} = {m_i}')
```

```python
n = random()
print(n, classify_sentiment(n))
```

## Matplotlib引入和使用

```python
# 定义x和y
x = list(range(10))
y = [i+1 for i in x]
print(x,y)
```

```python
from matplotlib import pyplot as plt
fig1 = plt.figure()
plt.plot(x,y)
plt.title('plot')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
fig2 = plt.figure()
plt.scatter(x,y)
plt.title('scatter')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

## numpy引入和使用

## Pandas引入和使用
